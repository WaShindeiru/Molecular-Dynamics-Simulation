```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from various import read_atoms_from_output_dumps, get_distance
from pathlib import Path
```

## c-c

```python
path_c = Path("/mnt/7E442D59442D1585/md/paper/pot_energy/c-c")

temp_c = read_atoms_from_output_dumps(path_c / "laamps", 0, 1)
```

```python
distance_c = get_distance(temp_c, 0, 1)

energy_data_c = pd.read_csv(path_c / "energy.csv", header=0)
```

```python
# plt.figure()
# plt.plot(distance['iteration'], distance['distance'])
# plt.show()

plt.figure()
plt.plot(distance_c['distance'], energy_data_c["potential_energy"], color='darkcyan', linewidth=3, label="Energia potencjalna")
plt.xlim([0, 4])
plt.ylim([-9, 20])
plt.xlabel("Odległość [A]", fontsize=15)
plt.ylabel("Energia potencjalna [eV]", fontsize=15)
plt.title("Zależność energii potencjalnej od odległości\n dla interakcji C-C", pad=5, fontsize=16)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=14)
plt.savefig('c-c-energy.png')
plt.show()
```
